## Data Preparation for Modeling


```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Separate features and target
X = df.drop(columns=['ID', 'Target'])
y = df['Target']
```

```python
# Split the data: 2260 rows for training and the last 400 rows for holdout testing
X_train = X.iloc[:-400, :]
y_train = y.iloc[:-400]
X_holdout = X.iloc[-400:, :]
y_holdout = y.iloc[-400:]
```

```python
# Impute missing values with 0
imputer = SimpleImputer(strategy='constant', fill_value=0)
X_train_imputed = imputer.fit_transform(X_train)
X_holdout_imputed = imputer.transform(X_holdout)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_holdout_scaled = scaler.transform(X_holdout_imputed)
```

##### Partial Dependence

```python
train_X_df = pd.DataFrame(X_train_imputed, columns=X.columns)
train_X_df.head()
```

```python
# Get sorted indices of the features
sorted_idx = perm_importance.importances_mean.argsort()

# Extract the top three feature indices
top_three_features_idx = sorted_idx[-3:]

# Get top three feature names
top_three_features = np.array(X.columns)[top_three_features_idx]

# Create partial dependence plots for the top three features
plt.figure(figsize=(12, 6))
for i, feature_idx in enumerate(top_three_features_idx):
    feature_name = X.columns[feature_idx]
    feature_values = np.linspace(X_holdout_scaled[:, feature_idx].min(), X_holdout_scaled[:, feature_idx].max(), num=50)
    partial_dependence = []

    # Calculate partial dependence for each feature value
    for value in feature_values:
        X_temp = X_holdout_scaled.copy()
        X_temp[:, feature_idx] = value
        partial_dependence.append(np.mean(random_forest_model.predict_proba(X_temp)[:, 1]))

    # Plot the partial dependence curve
    plt.subplot(1, 3, i + 1)
    plt.plot(feature_values, partial_dependence, label=feature_name)
    plt.xlabel(feature_name)
    plt.ylabel('Partial Dependence')
    plt.title(f'Partial Dependence Plot for {feature_name}')

plt.tight_layout()
plt.show()
```

```python
# Define the function for plotting a grid of ICE plots
def plot_ice_grid(ice_dfs, features, ax_ylabel, ncols=3, figsize=(15, 5), **ice_plot_kwargs):
    nrows = int(np.ceil(len(features) / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize)
    axes = axes.flatten()

    for i, feature in enumerate(features):
        ice_df = ice_dfs[feature]
        ax = axes[i]
        ice_plot(ice_df, ax=ax, **ice_plot_kwargs)
        ax.set_title(f'ICE Plot: {feature}')
        ax.set_ylabel(ax_ylabel)
        ax.set_xlabel(feature)

    # Remove any empty subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
# Select top 3 features
top_features = np.array(train_X_df.columns)[sorted_idx[-3:]]

# Create dict of ICE data for the top 3 features
train_ice_dfs = {feat: ice(data=train_X_df, column=feat, predict=lambda x: random_forest_model.predict_proba(x)[:, 1])
                 for feat in top_features}

# Plot ICE grid
plot_ice_grid(train_ice_dfs, top_features, ax_ylabel='Predicted Probability', alpha=0.3, plot_pdp=True,
              pdp_kwargs={'c': 'red', 'linewidth': 3}, linewidth=0.5, c='dimgray')
```

## Alternative Modeling Experiment

### Data Preparation

```python
# Preparing data
df.fillna(0, inplace=True)

X1 = df.drop(['ID','Target'], axis=1)
y1 = df['Target']
X1_train, X1_test = X1.iloc[:-400], X1.iloc[-400:]
y1_train, y1_test = y1.iloc[:-400], y1.iloc[-400:]

# SMOTE for handling data class imbalance
smote = SMOTE(random_state=42)
X1_train_resampled, y1_train_resampled = smote.fit_resample(X1_train, y1_train)

# Using Principle Component Analysis for dimension reduction
pca = PCA(n_components= 10)

# Scaling features
X1_train_scaled = pca.fit_transform(X1_train_resampled)
X1_test_scaled = pca.transform(X1_test)
```

### Modeling

```python
# Defining models to experiment/learning
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}

# Defining pipelines for each model
pipelines = {name: Pipeline([
        ('scaler', MinMaxScaler()),
        ('classifier', model)
    ]) for name, model in models.items()}
```

### Hyperparameter Tuning

```python
# Define hyperparameter grids for each model
param_grids = {
    "Random Forest": {'classifier__n_estimators': [100, 200, 400], 'classifier__max_depth': [5, 10, 15]},
    "Decision Tree": {'classifier__max_depth': [5, 10, 15]},
    "SVM": {'classifier__C': [0.1, 1, 10, 20, 50], 'classifier__gamma': ['scale', 'auto', 0.01, 0.1]},
    "Gradient Boosting": {'classifier__n_estimators': [100, 200, 300]}
}
```

### Evaluation and Metrics

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score

# Initializing variables to track the best model and F1 score
best_f1_score = 0
best_model_name = None
best_estimator = None
best_X1_test_scaled = None
best_y1_test = None
best_accuracy = 0
best_mae = None
best_mse = None
best_r2 = None

# Model training and evaluation with hyperparameter tuning using GridSearchCV for an exhaustive learning
for name, model in models.items():
    pipeline = Pipeline([
        ('scaler', MinMaxScaler()),
        ('classifier', model)
    ])

    grid_search = GridSearchCV(pipeline, param_grids[name], cv=5, scoring='f1')
    grid_search.fit(X1_train_scaled, y1_train_resampled)

    y1_pred = grid_search.predict(X1_test_scaled)
    current_f1_score = f1_score(y1_test, y1_pred)
    current_accuracy = accuracy_score(y1_test, y1_pred)
    current_mae = mean_absolute_error(y1_test, y1_pred)
    current_mse = mean_squared_error(y1_test, y1_pred)
    current_r2 = r2_score(y1_test, y1_pred)

    if current_f1_score > best_f1_score:
        best_f1_score = current_f1_score
        best_model_name = name
        best_estimator = grid_search.best_estimator_
        best_X1_test_scaled = X1_test_scaled
        best_y1_test = y1_test
        best_accuracy = current_accuracy
        best_mae = current_mae
        best_mse = current_mse
        best_r2 = current_r2

    # Results loop for our modeling techniques
    print(f"\nResults for {name}:")
    print("Best Parameters:", grid_search.best_params_)
    print("\nF1 Score:", current_f1_score)
    print("Accuracy:", current_accuracy)
    print("Mean Absolute Error:", current_mae)
    print("Mean Squared Error:", current_mse)
    print("R2 Score:", current_r2)
    print("\nClassification Report:")
    print(classification_report(y1_test, y1_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y1_test, y1_pred))

# Print best overall results
print("\nBest Model:")
print("Name:", best_model_name)
print("F1 Score:", best_f1_score)
print("Accuracy:", best_accuracy)
print("Mean Absolute Error:", best_mae)
print("Mean Squared Error:", best_mse)
print("R2 Score:", best_r2)
```
